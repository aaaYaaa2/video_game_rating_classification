````python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb
from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.metrics import accuracy_score as acc
from sklearn.tree import DecisionTreeClassifier 
from sklearn.linear_model import PassiveAggressiveClassifier 
from sklearn.linear_model import RidgeClassifier 
from sklearn.neighbors import KNeighborsClassifier
import joblib 

# read data 
video_game_rating_train = pd.read_csv("./data/Video_games_esrb_rating.csv")
video_game_rating_train.info()
````

````python
# EDA 
# look at the console distribution 
PS4_NUM = len(video_game_rating_train[video_game_rating_train['console'] == 0])
PS4_XBOX_NUM = len(video_game_rating_train[video_game_rating_train['console'] == 1])
labels = ['Play Station 4', 'Play Station 4 and XBOX']
values = [PS4_NUM, PS4_XBOX_NUM]

plt.figure(figsize=(17, 5))
plt.subplot(1,2,1)
plt.pie(values, labels = labels, autopct='%1.2f%%')
plt.title('Pie Chart: Console distribution of Video Games Rating Dataset')
plt.show()
````

````python
# rating distribution 
uniqueRatingVals = set(video_game_rating_train['esrb_rating'])
countArr = []
for rate_name in uniqueRatingVals:
    countArr.append(len(video_game_rating_train[video_game_rating_train['esrb_rating'] == rate_name]))
    
plt.figure(figsize=(17, 5))
plt.subplot(1, 2, 1)
plt.bar(list(uniqueRatingVals), countArr, color = 'orange')
plt.title(' Bar graph: Rating Distribution ')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.pie(countArr, labels = list(uniqueRatingVals), autopct='%1.2f%%')
plt.title(' Pie chart: Rating Distribution ')
plt.show()
````

````python
# relabel esrb_rating to numbers 
rating_mapping = {
    'E': 0,
    'ET': 1,
    'T': 2,
    'M': 3
}

video_game_rating_train['esrb_rating'] = video_game_rating_train['esrb_rating'].map(rating_mapping)
````

````python
'''
    Looking at the dataset, we could see that there are some features that might be correlated 
    with each other. 
    
    For instances: animated_blood, blood, blood_and_gore, mild_blood are all related to blood
    violent type. 
    To detect those correlations, we will plot a correlation map between different features 
'''
plt.figure(figsize = (30, 30))
featuresRequired = video_game_rating_train.columns[2:]
sb.heatmap(video_game_rating_train[featuresRequired].corr(), annot = True)
plt.show()
````

````python
'''
There are several features with high correlation 
[intense_violence, blood_and_gore] 0.29
[strong_language, blood] 0.29
[strong_sexual_content, blood_and_gore] 0.34
[strong_language, blood_and_gore] 0.22
[simulated_gambling, crude_humor] 0.32
[strong_secual_content, intense_violence] 0.39
[strong_language, sexual_themes] 0.27
[mild_blood, fantasy_violence] 0.23

most of the correlated feature are blood related features 
'''
```

````

````python
# split the set into training and testing
# remove the first column, title 
video_game_model_dataset = video_game_rating_train.loc[:, 'console':'esrb_rating']
video_game_model_dataset.info()
X_train, X_test, y_train, y_test = train_test_split(
    video_game_model_dataset.values[:,:-1],
    video_game_model_dataset.values[:,-1:],
    test_size=0.25,
    random_state=42
)

y_train = y_train.ravel()
y_test = y_test.ravel()
print('Training Dataset Shape: ', X_train.shape, y_train.shape)
print('Testing Dataset Shape: ', X_test.shape, y_test.shape)
````

````python
# 1. Classifier: Random forest 
random_forest_classifier = RandomForestClassifier(n_estimators = 100, n_jobs=-1)
# step forward feature selection 
sfs1 = sfs(
    random_forest_classifier,
    k_features = 20,
    forward=True,
    floating=False,
    verbose=2,
    scoring='accuracy',
    cv=5
)

# Perform SFFS 
sfs1 = sfs1.fit(X_train, y_train)
````

````python
# find the feature 
feat_cols = list(sfs1.k_feature_idx_)
print(feat_cols)
````

````python
# build the full model with selected features 
clf = RandomForestClassifier(n_estimators=100, random_state=42, max_depth=4)
clf.fit(X_train[:, feat_cols], y_train)

y_train_pred = clf.predict(X_train[:, feat_cols])
y_test_pred = clf.predict(X_test[:, feat_cols])
print('Training accuracy on selected features: %.3f' % acc(y_train, y_train_pred))
print('Testing accuracy on selected features: %.3f' % acc(y_test, y_test_pred))
````

````python
# Model 2 Decision Tree Model 
decision_tree_classifier = DecisionTreeClassifier(random_state=50)

# step forward feature selection 
sfs2 = sfs(
    decision_tree_classifier,
    k_features = 20,
    forward=False,
    floating=False,
    verbose=2,
    scoring='accuracy',
    cv=5
)

# Perform SFFS on Decision Tree Model  
sfs2 = sfs2.fit(X_train, y_train)
````

````python
feat_cols_decision_tree = list(sfs2.k_feature_idx_)
print(feat_cols_decision_tree)

dtc = DecisionTreeClassifier(random_state=50)
dtc.fit(X_train[:, feat_cols_decision_tree], y_train)

y_train_pred_decision_tree = dtc.predict(X_train[:, feat_cols_decision_tree])
y_test_pred_decision_tree = dtc.predict(X_test[:, feat_cols_decision_tree])
print('Training accuracy on selected features: %.3f' % acc(y_train, y_train_pred_decision_tree))
print('Testing accuracy on selected features: %.3f' % acc(y_test, y_test_pred_decision_tree))
````

````python
# Model 3 Passive Aggressive Classifier
passive_aggressive_classifier = PassiveAggressiveClassifier(random_state=45)
sfs3 = sfs(
    passive_aggressive_classifier,
    k_features = 20,
    forward=True,
    floating=False,
    verbose=2,
    scoring='accuracy',
    cv=5
)

# Perform SFFS on Decision Tree Model  
sfs3 = sfs3.fit(X_train, y_train)
````

````python
feat_cols_passive_aggressive = list(sfs3.k_feature_idx_)
print(feat_cols_passive_aggressive)

pac = PassiveAggressiveClassifier(random_state=45)
pac.fit(X_train[:, feat_cols_passive_aggressive], y_train)

y_train_pred_passive_aggressive = pac.predict(X_train[:, feat_cols_passive_aggressive])
y_test_pred_passive_aggressive = pac.predict(X_test[:, feat_cols_passive_aggressive])
print('Training accuracy on selected features: %.3f' % acc(y_train, y_train_pred_passive_aggressive))
print('Testing accuracy on selected features: %.3f' % acc(y_test, y_test_pred_passive_aggressive))
````

````python
# Model 4 Ridge Classifier 
ridge_classifier = RidgeClassifier(random_state=45)
sfs4 = sfs(
    ridge_classifier,
    k_features = 20,
    forward=True,
    floating=False,
    verbose=2,
    scoring='accuracy',
    cv=5
)

# Perform SFFS on Decision Tree Model  
sfs4 = sfs4.fit(X_train, y_train)
````

````python
feat_cols_ridge = list(sfs4.k_feature_idx_)
print(feat_cols_ridge)

pac = RidgeClassifier(random_state=45)
pac.fit(X_train[:, feat_cols_passive_aggressive], y_train)

y_train_pred_passive_aggressive = pac.predict(X_train[:, feat_cols_passive_aggressive])
y_test_pred_passive_aggressive = pac.predict(X_test[:, feat_cols_passive_aggressive])
print('Training accuracy on selected features: %.3f' % acc(y_train, y_train_pred_passive_aggressive))
print('Testing accuracy on selected features: %.3f' % acc(y_test, y_test_pred_passive_aggressive))

````

````python
kneighbors_classifier = KNeighborsClassifier(n_neighbors=4)
sfs5 = sfs(
    kneighbors_classifier,
    k_features = 20,
    forward=True,
    floating=False,
    verbose=2,
    scoring='accuracy',
    cv=5
)

# Perform SFFS on Decision Tree Model  
sfs5 = sfs5.fit(X_train, y_train)
````

````python
feat_cols_kneighbors = list(sfs5.k_feature_idx_)
print(feat_cols_kneighbors)

kneighbors = KNeighborsClassifier(n_neighbors=4)
kneighbors.fit(X_train[:, feat_cols_kneighbors], y_train)

y_train_pred_kneighbors = kneighbors.predict(X_train[:, feat_cols_kneighbors])
y_test_pred_kneighbors = kneighbors.predict(X_test[:, feat_cols_kneighbors])
print('Training accuracy on selected features: %.3f' % acc(y_train, y_train_pred_kneighbors))
print('Testing accuracy on selected features: %.3f' % acc(y_test, y_test_pred_kneighbors))
````
